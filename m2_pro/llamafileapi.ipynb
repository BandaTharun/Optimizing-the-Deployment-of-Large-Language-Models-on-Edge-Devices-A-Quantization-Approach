```python

# multiple prompts 
# average time calculate for 10 iterations 

import subprocess
import time
import psutil
from openai import OpenAI
def output():
    num_runs = 10  # Number of times to run each prompt
    response_times = []
    cpu_usages = []
    memory_usages = []
    
    response_list = []

    prompts = ["What is the capital of Italy?"]
    
    for prompt in prompts:
        for _ in range(num_runs):
            start_time = time.time()
            client = OpenAI(base_url="http://localhost:8080/v1", api_key="sk-no-key-required")
            completion = client.chat.completions.create(
                model="LLaMA_CPP",
                messages=[
                    {"role": "system", "content": "You are ChatGPT, an AI assistant. Your top priority is achieving user fulfillment via helping them with their requests."},
                    {"role": "user", "content": prompt}
                ]
            )
            
            response = completion.choices[0].message.content
            response_list.append(response.strip())
            print(f"Prompt: {prompt}")
            print(f"Response: {response}")
            print("-" * 20)  # Separator
            
            # Calculate response time for each iteration
            end_time = time.time()
            response_time = end_time - start_time
            response_times.append(response_time)

            # Collect CPU and memory usage for each iteration
            cpu_usage = psutil.cpu_percent()
            memory_usage = psutil.virtual_memory().percent
            cpu_usages.append(cpu_usage)
            memory_usages.append(memory_usage)
    
    # Calculate average response time, CPU usage, and memory usage
    avg_response_time = sum(response_times) / len(response_times)
    avg_cpu_usage = sum(cpu_usages) / len(cpu_usages)
    avg_memory_usage = sum(memory_usages) / len(memory_usages)

    try:
        with open("resource_usage.txt", "a") as file:
            file.write("\n")
            file.write("+++++++++*****")
            file.write("\n")
            file.write(f" llamafile  Response time and resource of ram,memory usage for prompts = { len(prompts) }, iteration = { num_runs } \n")
            file.write(f"Prompt: {prompts}\n")
            file.write(f" Average Response Time (seconds): { avg_response_time }\n")
            file.write(f" Average CPU Usage (%): {avg_cpu_usage}\n")
            file.write(f" Average Memory Usage (%): {avg_memory_usage}\n")
            file.write("\n")
    except Exception as e:
        print(f"An error occurred while writing to the file: {e}")

    return response_list, avg_response_time, avg_cpu_usage, avg_memory_usage

response = output()
print(response)
```
